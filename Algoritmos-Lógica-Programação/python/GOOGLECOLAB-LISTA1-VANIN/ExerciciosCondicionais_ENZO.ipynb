## **Exemplo 01**
A empresa *XSoftwares* concedeu um bônus de 20% do valor do salário a todos os
funcionários com tempo de trabalho na empresa igual ou superior a cinco anos, e
10% aos demais. Calcular e exibir o valor do bônus.
Para resolver o problema é necessário conhecer o valor do salário e o tempo
de serviço do funcionário.

### **Algoritmo em Visualg**
image.png

```python
# ENTRADA DE DADOS E ATRIBUIÇÃO PARA AS RESPECTIVAS VARIÁVEIS
salario = float(input("Digite o salário do funcionário:"))
tempo = int(input("Digite a quantidade de horas trabalhadas:"))

#PROCESSAMENTO
if(tempo>=5):
  bonus = float(salario*0.20)
else:
  bonus = float(salario*0.10)

#SAIDA
print("O valor do bonus é de R$",round(bonus,2))
```

## **Exemplo 02**
Sabemos que um aluno é aprovado caso apresente média maior ou igual a 7.0 e
frequência maior ou igual a 75%. Na verdade, em uma situação real, se o aluno
obtiver a frequência mínima exigida e uma média entre 3 e 7, ainda teria direito a uma última avaliação de recuperação.

### **Algoritmo em Visualg**
![image.png](EXERCICIO1_PRINT.)

```python
# SOLUÇÃO 01
media=float(input("Digite a média do aluno:"))
frequencia=float(input("Digite a frequencia do aluno:"))

if(frequencia>=75):
  if(media>=7):
    print("APROVADO")
  else:
    if(media>=3):
      print("RECUPERAÇÃO")
    else:
      print("REPROVADO POR MÉDIA")
else:
  print("REPROVADO POR FALTAS")
```

```python
# SOLUÇÃO 02
media=float(input("Digite a média do aluno:"))
frequencia=float(input("Digite a frequencia do aluno:"))

if(frequencia>=75 and media>=7):
  print("APROVADO")
elif(frequencia>=75 and media<3):
  print("REPROVADO POR MÉDIA")
elif(frequencia>=75 and media>=3 and media<7):
  print("RECUPERAÇÃO")
else:
  print("REPROVADO POR FALTAS")
```

Para o exercício 06 considerar a fórmula para se obter o peso ideal:
Masculino ==> PesoIdeal = 52 + (0.75 × (altura - 152.4)
Feminino ==>  PesoIdeal = 52 + (0.67 × (altura - 152.4)

```python
sexo = input('Digite seu sexo (masculino/feminino): ')
altura = float(input('Digite sua altura em centímetros: '))
peso = float(input('Digite seu peso em quilogramas: '))

if sexo == 'masculino':
    pesoH = 52 + (0.75 * (altura - 152.4))
    print('Seu peso atualmente é de: ', peso)
    print('Seu peso ideal é de: ', pesoH)
    if peso > pesoH:
        print('Seu peso está acima do ideal')
    elif peso < pesoH:
        print('Seu peso está abaixo do ideal')
    else:
        print('Seu peso está dentro do ideal')

elif sexo == 'feminino':
    pesoM = 52 + (0.67 * (altura - 152.4))
    print('Seu peso atualmente é de: ', peso)
    print('Seu peso ideal é de: ', pesoM)
    if peso > pesoM:
        print('Seu peso está acima do ideal')
    elif peso < pesoM:
        print('Seu peso está abaixo do ideal')
    else:
        print('Seu peso está dentro do ideal')


```
